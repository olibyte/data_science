## Segment 2 - A neural network with a Perceptron

```python
import numpy as np
import pandas as pd
import sklearn

from pandas import Series, DataFrame
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.linear_model import Perceptron
```

```python
iris = datasets.load_iris()

X = iris.data
y = iris.target

X[0:10,]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#normalize the data
standardize = StandardScaler() #instantiate standardscaler object

standardized_X_test = standardize.fit_transform(X_test)

standardized_X_train = standardize.fit_transform(X_train)
```

```python
standardized_X_test[0:10,]
```

```python
perceptron = Perceptron(max_iter=50, eta0=0.15, tol=1e-3, random_state=15) #random state to set the seed

perceptron.fit(standardized_X_train, y_train.ravel()) #ravel to format properly
```

```python
#make a prediction
y_pred = perceptron.predict(standardized_X_test)
```

```python
print(y_test)
```

```python
print(y_pred)
```

```python
print(classification_report(y_test, y_pred))
#check accuracy of model's predictions
```
