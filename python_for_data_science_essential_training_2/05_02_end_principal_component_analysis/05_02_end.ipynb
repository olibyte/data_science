### PCA on the iris dataset

```python
iris = datasets.load_iris()
X = iris.data
variable_names = iris.feature_names

X[0:10,]
```

```python
#instantiate PCA object
pca = decomposition.PCA()
#perform dimension reduction
iris_pca = pca.fit_transform(X)

pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
#EVR tells us how much info is compressed into first few components
#When deciding how many components to keep, look at % of cumulative variance. Retain at least 70% of dataset's original info
#look at principle components df
comps = pd.DataFrame(pca.components_, columns=variable_names)
comps
```

```python
sb.heatmap(comps, cmap="Blues", annot=True) #correlation heatmap to see how vars correlate with principal components data
#component 1 = row 0. strongly positively correlated w/ petal length, moderately w/ petal width and sepal length, etc
```
